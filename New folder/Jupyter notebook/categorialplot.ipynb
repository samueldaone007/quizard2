```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('supermarket_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize= (10,5))
sns.countplot(x='Product line',hue='Gender', data=df, palette='pink')
plt.xticks(rotation = 90)
```

```python
sns.boxplot(x = 'Product line',y = 'Total', hue='Customer type', data=df)
plt.xticks(rotation = 90)
```

```python
sns.violinplot(x='Customer type', y='Total', hue='Gender', data=df)
plt.xticks(rotation = 90)
```

```python
plt.figure(figure=(10,5))
sns.stripplot(x='Payment', y='Total', hue='City', data=df)
plt.legend(loc = 'upper right')
```

```python
sns.pairplot(df, hue='City')
```

```python
sns.scatterplot(x='cogs',y='Unit price', data=df)
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['bill_length_mm'].fillna(df['bill_length_mm'].mean(), inplace=True)
```

```python
df['body_mass_g'].fillna(df['body_mass_g'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df['sex'].mode()
```

```python
df['sex'].fillna(df['sex'].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='body_mass_g', y='flipper_length_mm',hue='species',data=df)
plt.legend(loc = 4)
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='body_mass_g', y='bill_depth_mm',hue='species',data=df)
plt.legend(loc = 'upper right')
```
