```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
#swarmplot
sns.swarmplot(x='day', y='total_bill', data=df, hue='sex')
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```
