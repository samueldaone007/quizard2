```python
data = sns.load_dataset('penguins')
```

```python
billDepth = data['bill_depth_mm'].mean()
```

```python
billDepth
```

```python
data['bill_depth_mm'].fillna(billDepth, inplace = True)
```

```python
bill_length = data['bill_length_mm'].mean()

data['bill_length_mm'].fillna(bill_length, inplace = True)
```

```python
flipper_length = data['flipper_length_mm'].mean()

data['flipper_length_mm'].fillna(flipper_length, inplace = True)
```

```python
data.isna().sum()
```

```python
bmass = data['body_mass_g'].mean()

data['body_mass_g'].fillna(bmass, inplace=True)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
modeS = data['sex'].mode()
```

```python
modeS
```
