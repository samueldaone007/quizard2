```python
def new_column(quality):
    if quality >= 3 and quality <=4:
        return 'low'
    elif quality >= 5 and quality <=6:
        return 'medium'
    elif quality >= 7 and quality <=8:
        return 'high'
    else:
        return 'very high'
df['Quality_new'] = df['quality'].apply(new_column)
```

```python
df.value_counts('Quality_new')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x = df.drop(df[['quality', 'Quality_new', 'Id']], axis = 1)
```

```python
y = df['Quality_new']
```

```python
encoder = LabelEncoder()
```

```python
encodedy = encoder.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,encodedy, test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.score(x_train,y_train)
```

```python
infere = classifier.predict(x_test)
```

```python
infere
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(infere,y_test)
```
